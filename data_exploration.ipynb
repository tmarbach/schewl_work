# Snake Data Exploration
Prior to working with the dataset an initial exploration was done to understand what was availible in the dataset. 

```python
import pandas as pd
from window_maker import leaping_window
import matplotlib.pyplot as plt
from dataset_defintions import *
from clean_acceleration_data import remove_case_sensitivity
```

```python
def concat_data(behavior_windows):
    return pd.concat(behavior_windows).reset_index()

def describe_data(behavior_windows):
    return behavior_windows[COLUMN_NAMES].describe()

def create_behavior_figure(behavior_windows, title):
    plt.rcParams["figure.figsize"] = (20,4)
    plt.plot(behavior_windows.index, behavior_windows[ACCELERATION_X], color='r', label='Line x')
    plt.plot(behavior_windows.index, behavior_windows[ACCELERATION_Y], color='g', label='Line y')
    plt.plot(behavior_windows.index, behavior_windows[ACCELERATION_Z], color='b', label='Line z')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
PATH = "./dataset_subset/"
DATA_SETS = [str(PATH + '2021-10-05_oreganus_Pen11__083121_41.csv'),
             str(PATH + '2021-10-05-oreganus_Pen11_051721_4.csv'),
             str(PATH + '2021-10-21_oreganus_JLTA_092021_10(defensive).csv'),
             str(PATH + '2021-10-21_oreganus_Pen10_092021_10(defensive).csv') ]

df = pd.concat([pd.read_csv(csv, low_memory=False) for csv in DATA_SETS]).reset_index()
df = remove_case_sensitivity(df)
```

# Defensive Strike (hit)

```python
DEFENSIVE_STRIKE_HIT_LABEL = "Defensive Strike (hit)"
behavior_windows, class_list = leaping_window(df, WINDOW_SIZE, ['h'])

if(len(class_list) > 0):
    behavior_windows = concat_data(behavior_windows)
    print(describe_data(behavior_windows))
    create_behavior_figure(behavior_windows, DEFENSIVE_STRIKE_HIT_LABEL)
else:
    print(f"The dataset was unable to form an adaquate amount of windows for the {DEFENSIVE_STRIKE_HIT_LABEL} behavior")
```
