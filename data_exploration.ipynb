```python
import pandas as pd
from window_maker import pull_window
import matplotlib.pyplot as plt

# Two CSV's has a behavior column that has no label in the header
# Swallow, Strike CSV.
# Hit and Miss are only in the defensive CSV.
PATH = "./data_subsets"
DATA_SETS = [PATH + '2021-10-21_oreganus_JLTA_092021_10(defensive).csv',
             PATH + '2021-10-21_oreganus_JLTA_092021_10.csv']

df = pd.concat([pd.read_csv(csv, low_memory=False) for csv in DATA_SETS]).reset_index()
df.to_feather('total.feather')
```

```python
# Pull window for all of these.
# Or at least just showcase x,y,z
class_list = ['c', 'i', 'w', 's','l','t', 'a','d']
sdf, clist = pull_window(df, 25, ['s'])
sdf = pd.concat(sdf).reset_index()
```

```python
pf = pd.read_csv("~/CNNworkspace/total_clean_data.csv")
pf = pf.rename(columns={"Behavior":"behavior"})
```

```python
tf, clist = pull_window(pf, 25, ['d'])
tf = pd.concat(tf).reset_index()
plt.rcParams["figure.figsize"] = (20,4)
plt.plot(tf.index, tf['accX'], color='r', label='Line x')
plt.plot(tf.index, tf['accY'], color='g', label='Line y')
plt.plot(tf.index, tf['accZ'], color='b', label='Line z')
plt.title("Drag Carcass")
plt.legend()
plt.show()
```
